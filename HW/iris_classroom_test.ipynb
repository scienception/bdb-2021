```python
import pandas as pd
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
# How many rows does the data set contain? How many columns?
print("col", iris.shape[1])
print("row", iris.shape[0])
```

```python
# Another way:
row, col = iris.shape
print("There are %s colums and %s rows" %(col, row))
```

```python
# Compute the average petal length.
iris['petal_length'].mean()
```

```python
# mean of all the columns
iris.mean()
```

```python
# Find the rows of the data set where petal length is 50% longer 
# than the average petal length.
m = iris['petal_length'].mean()
small_df = iris.query('petal_length > %f' % (1.5*m))
small_df.head()
```

```python
# Find the average sepal width for each species.

iris_species = iris.groupby('species')
iris_species['sepal_width'].mean()
```

```python
# in one line
iris.groupby('species')['sepal_width'].mean()
```

```python
# Plot the Iris data set with a scatter plot (use kind='scatter' 
# when calling the plot() method), plotting petal_length vs. 
# petal_width.

iris.plot(x='petal_length', y='petal_width', kind='scatter')
```

```python
# Repeat the plot above but this time assign different colors to the
# different Iris species, which should all be plotted in the same 
# plot. Hint: you will need to find some suitable parameters for the
# plot() method. Note: it is not necessary to use toolkits other
# than Pandas to do this.

iris.query('species == "setosa"')
```

```python
iris.query('species == "setosa"').plot(x='petal_length', y='petal_width', kind='scatter', color='r')
iris.query('species == "versicolor"').plot(x='petal_length', y='petal_width', kind='scatter', color='b')
iris.query('species == "virginica"').plot(x='petal_length', y='petal_width', kind='scatter', color='g')
```

```python
# how to plot them together
ax = iris.query('species == "setosa"').plot(x='petal_length', y='petal_width', kind='scatter', color='r', label = 'setosa')
iris.query('species == "versicolor"').plot(x='petal_length', y='petal_width', kind='scatter', color='b', ax=ax,  label = 'versicolor')
iris.query('species == "virginica"').plot(x='petal_length', y='petal_width', kind='scatter', color='g', ax=ax,  label = 'virginica')
```
